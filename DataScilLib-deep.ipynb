####  fetch columns with different methods iloc, loc, ix

 * loc gets rows (or columns) with particular labels from the index.
 * iloc gets rows (or columns) at particular positions in the index (so it only takes integers).
 * ix usually tries to behave like loc but falls back to behaving like iloc if a label is not present in the index.

```python
# fetch the row which has index no as 3

print("row 3 :\n",df.loc[3])

# fetch the row 2:3
print("\n\n row 2 , 3 :\n",df.iloc[1:3])


# we can also select columumns with rows select only name and age of TOM 
print("\n\n row and colum : \n",df.iloc[2,1:3])



# check datatype 
print("\nDatatype : ",df.dtypes)
```

## Merging in Pandas

```python
## to merge dataframe we have to pass the parameter that how we want to merge , on which column we 
# want to merge 
# in merge the one attribute has to be same in both dataframes

df1 = {"Name":["John","Messi","Tom","SRK","Rock","Vin Diesel"],
        "Age":[40,32,45,46,45,40],
       "City":["US","Argentina","US","India","US","UK"]}

df2 = {"Name":["John","Messi","Tom","SRK","Ronaldo"],
        "Profession":["wrestling","Athelte","Actor","Actor","Athlete"]}

df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)

#Inner merge
print(" Inner Merge:  \n",df1.merge(df2,on='Name',how='inner'))

 
#Outer merge
print(" \n\nRight Merge:  \n",df1.merge(df2,on='Name',how='right'))

#Left merge
print(" \n\nLeft Merge:  \n",df1.merge(df2,on='Name',how='left'))


#outer join
print(" \n\nOuter Merge:  \n",df1.merge(df2,on='Name',how='outer'))

```

## Join operation in Pandas

```python
## Join is same work as merge bit the difference is that it join the datframe
# on the base of commmon index name

# * Both dataframe must have different attributes names

df1 = {"Name":["John","Messi","Tom"],
        "Age":[40,32,45],
       "City":["US","Argentina","US"]}

df2 = {"Names":["John","Messi","Tom","SRK"],
        "Profession":["wrestling","Athelte","Actor","Actor"]}

df1 = pd.DataFrame(df1, index = ['I1','I2','I3'])
df2 = pd.DataFrame(df2, index = ['I1','I2','I3','I8'])

#inner join
print("Inner join\n",df1.join(df2,how="inner"))

#outer join
print("\n\nOuter join\n",df1.join(df2,how="outer"))

#left join
print("\n\nLeft join\n",df1.join(df2,how="left"))

#right join
print("\n\n Right join\n",df1.join(df2,how="right"))
```
