```python
####    MEAN, median and MOde function



#because we need geometric mean 

from scipy import stats

data_main = {"Name":['Faiz','Shish','Ahesan','Raj','Mayank','Bhavik'],"En_no":[1027,1001,1001,1105,1060,1008],
             "city":['palanpur','Bhuj','palanpur','kathiyavad','kathiyavad','Khedbramha']}

datamain1={"Score1":[102,156,145,189,120],"Score2":[102,156,145,200,120],"Score3":[102,156,145,189,120]}
df1=pd.DataFrame(datamain1)
df = pd.DataFrame(data_main)
data = df['En_no']
print("Mean is ",data.mean(),end="\n\n")

print("Median is",data.median(),end="\n\n")

print("mode is",data.mode(),end="\n\n")

print("gmean is",stats.gmean(data),end="\n\n")
a=len(df)

print("Std is", data.std(),end="\n\n\n")

print("Variance",df1.var(),end="\n\n\n")
print
#now suppose i want to assign the value of mode to entire column
for i in range (0,a):
    df['En_no'][i]=data.mode()

print(df)
```

```python
### Reshaping in pandas 
#melt() function is useful to massage a DataFrame into a format where one or more columns are identifier variables, while all other columns, considered measured variables, are unpivoted to the row axis, leaving just two non-identifier columns, variable and value.

import pandas as pd

data_main = {"Name":['Faiz','Faiz','Shish','Ahesan','Raj','Mayank','Bhavik'],"En_no":[1027,1027,1001,1001,1105,1060,1008],
             "sex":['Male','Male','Female','Female','Female','Male','Female'],
             "city":['palanpur','palanpur','Bhuj','palanpur','kathiyavad','kathiyavad','palanpur'],
             "job":['DataScientist','DataScientist','Developer','HR','CEO','HR','Developer']}


df = pd.DataFrame(data_main)
print(df,end="\n\n")
df = pd.melt(df,id_vars=["sex"],var_name="Name",value_name="temprature")
df
#print(df[df["sex"]=="Male"])
```

```python
# Pivot and Pivot table
data_main = {"Name":['Faiz','Faiz','Shish','Ahesan','Raj','Mayank','Bhavik'],"En_no":[1027,1027,1001,1001,1105,1060,1008],
             "sex":['Male','Male','Female','Female','Female','Male','Female'],
             "city":['palanpur','palanpur','Bhuj','palanpur','kathiyavad','kathiyavad','palanpur'],
             "job":['DataScientists','DataScientists','Developer','HR','CEO','HR','Developer']}


## pivot ( index, columns, and values )
df = pd.DataFrame(data_main)

df = df.drop_duplicates()

df.set_index("En_no")
df= df.pivot(index="Name",columns="sex",values="city")
df
```
